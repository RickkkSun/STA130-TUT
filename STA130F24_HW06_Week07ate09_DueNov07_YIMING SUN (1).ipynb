# Q1.

In a Simple Linear Regression model, we aim to describe the relationship between an outcome variable Y (often called the dependent or response variable) and a predictor variable X (the independent variable). Mathematically, we represent this relationship as:

**Yi​=β0​+β1​X i​+ϵi​**

where:

* 𝛽0 is the intercept, representing the expected value of Y when X=0.
* 𝛽1 is the slope, indicating the change in Y for a one-unit change in X.
* 𝜖𝑖 is the error term (or noise), assumed to be independently drawn from a normal distribution with mean 0 and standard deviation σ, capturing the random variations that are not explained by the linear relationship.

The predictors 𝑋𝑖 can be chosen or sampled independently, such as from a uniform distribution, to form a set of fixed values. We generate a sample of Y values that approximate a normal distribution around the regression line.

````python
import numpy as np
import scipy.stats as stats
import plotly.graph_objs as go

# Parameters
n = 100  # Number of observations
beta0 = 2.0  # Intercept
beta1 = 1.5  # Slope
sigma = 1.0  # Standard deviation of the error term

# Generate predictor variable X from a uniform distribution
X = stats.uniform.rvs(0, 10, size=n)  # X values between 0 and 10

# Generate random error term epsilon from a normal distribution
epsilon = np.random.normal(0, sigma, n)

# Calculate outcome variable Y
Y = beta0 + beta1 * X + epsilon

# Create the regression line without noise
Y_line = beta0 + beta1 * X

# Plot the regression line and the data points
trace_data = go.Scatter(x=X, y=Y, mode='markers', name='Simulated Data Points')
trace_line = go.Scatter(x=X, y=Y_line, mode='lines', name='Regression Line')

layout = go.Layout(
    title='Simple Linear Regression Model Simulation',
    xaxis=dict(title='X (Predictor Variable)'),
    yaxis=dict(title='Y (Outcome Variable)')
)

fig = go.Figure(data=[trace_data, trace_line], layout=layout)
fig.show(renderer="png")
````

# Q2.

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.graph_objs as go

# Parameters for the theoretical model
n = 100  # Number of observations
beta0 = 2.0  # Intercept
beta1 = 1.5  # Slope
sigma = 1.0  # Standard deviation of the error term

# Generate predictor variable X from a uniform distribution
X = np.random.uniform(0, 10, n)  # X values between 0 and 10

# Generate random error term epsilon from a normal distribution
epsilon = np.random.normal(0, sigma, n)

# Calculate outcome variable Y based on the theoretical model
Y = beta0 + beta1 * X + epsilon

# Create a pandas DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit a Simple Linear Regression model using statsmodels
model = smf.ols('Y ~ X', data=data).fit()

# Print the model summary
print(model.summary())

# Generate the fitted values for visualization
data['Y_pred'] = model.predict(data['X'])

# Plot the observed data points and the fitted regression line
trace_data = go.Scatter(x=data['X'], y=data['Y'], mode='markers', name='Observed Data')
trace_fit = go.Scatter(x=data['X'], y=data['Y_pred'], mode='lines', name='Fitted Regression Line')

layout = go.Layout(
    title='Fitted Simple Linear Regression Model',
    xaxis=dict(title='X (Predictor Variable)'),
    yaxis=dict(title='Y (Outcome Variable)')
)

fig = go.Figure(data=[trace_data, trace_fit], layout=layout)
fig.show(renderer="png")
````

**Firstly, the fitted_model.fittedvalues in a Simple Linear Regression model are the predicted values for each observation in the dataset. These predicted values are calculated directly from the estimated coefficients for the intercept and slope, which are stored in fitted_model.params.**

From fomular itself(**Yi​=β0​+β1​X i​+ϵi​**), I have explored a little bit in Q1, however:

**Fitted Model: In practice, we don’t know the true values of β0and β1. Instead, we estimate them from sample data, giving us estimated intercept and estimated slop. Using these estimates, the fitted model for each observation 𝑖 is:** $\hat{y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i$


Here are three steps which can be derived from my basic understanding:

1. Estimated Coefficients: After fitting the regression model, statsmodels provides the estimated coefficients in fitted_model.params. This object contains:

* fitted_model.params[0] (or fitted_model.params['Intercept']), which is the estimated intercept.
* fitted_model.params[1] (or fitted_model.params['X']), which is the estimated slope for the predictor variable X.

2. Prediction Calculation: fitted_model.fittedvalues calculates the predicted values by applying these coefficients to each observation's predictor value X. For each observation in the dataset, it multiplies the predictor value by the estimated slope and adds the estimated intercept.

3. Summary Table Reference: If I examine fitted_model.summary().tables[1], I'll see these coefficients in the "coef" column, which directly correspond to fitted_model.params. These values are then used to calculate fitted_model.fittedvalues for each observation.

Simply summarize, fitted_model.fittedvalues is derived by applying **the estimated intercept and slope (from fitted_model.params) to each observation’s predictor variable, providing the model's in-sample predictions based on the fitted regression line.**

**The line chosen for the fitted model using the Ordinary Least Squares (OLS) method is the one that minimizes the total squared differences between the observed data points and the predicted values on the line. This line is often referred to as the "best-fit" line.**

1. Preventing Cancellation: Squaring ensures that positive and negative residuals don’t cancel each other out, giving a true measure of total error.

2. Penalizing Larger Errors: Squaring emphasizes larger errors more, which helps ensure the line is as close as possible to all points.

By minimizing this sum of squared errors, OLS provides the most accurate line through the data based on observed patterns, balancing overall prediction accuracy across all points.

````python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

# Set parameters for simulation
n, x_min, x_range, beta0, beta1, sigma = 20, 5, 5, 2, 3, 5
x = stats.uniform(x_min, x_range).rvs(size=n)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)
Y = beta0 + beta1 * x + errors

# Create DataFrame
df = pd.DataFrame({'x': x, 'y': Y})

# Fit the model
model_data_specification = smf.ols("y ~ x", data=df) 
fitted_model = model_data_specification.fit() 

# Add a column for color-coding in the plot
df['Data'] = 'Data'  # hack to add data to legend

# Create scatter plot with trendline
fig = px.scatter(df, x='x', y='y', color='Data', 
                 trendline='ols', title='y vs. x')

# Overlay the trendline from statsmodels
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

# Define the theoretical line based on beta0 and beta1
x_range = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range
fig.add_scatter(x=x_range, y=y_line, mode='lines',
                name=str(beta0) + ' + ' + str(beta1) + ' * x', 
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],
                    y=[fitted_model.fittedvalues[i], df['y'][i]],
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)
    
# Add horizontal line at y-bar
fig.add_scatter(x=x_range, y=[df['y'].mean()] * 2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

# Show plot
fig.show(renderer="png")
````

Interpretation of the Expressions for R-squared:

1. The first expression: 1-((Y-fitted_model.fittedvalues)**2).sum()/((Y-Y.mean())**2).sum()

Total Variation in Y:

The denominator in this expression, ((Y - Y.mean())**2).sum(), represents the total variation in Y. This measures how much the values of Y differ from their overall average and captures the overall spread of Y values around their mean.
Unexplained Variation (Residual Sum of Squares):

2. The numerator, ((Y - fitted_model.fittedvalues)**2).sum(), 

calculates the unexplained variation, or residual sum of squares. This is the sum of the squared differences between the actual Y values and the model’s predicted values, fitted_model.fittedvalues. This residual variation measures how far off the model's predictions are from the observed values of Y.
Proportion of Variation Explained:

Subtracting the ratio of unexplained variation from 1 gives us the proportion of the total variation in Y that is explained by the model. This value is called R-squared. It shows how well the model captures the variability in the data: a higher R-squared value means the model explains more of the variation in Y.
Therefore, fitted_model.rsquared in statsmodels directly provides this proportion of variation in Y that the model explains. This value serves as a measure of the model’s accuracy: the closer R-squared is to 1, the more accurately the model captures the data’s trend.

3. Interpretation of np.corrcoef(Y, fitted_model.fittedvalues)[0,1]**2

The expression np.corrcoef(Y, fitted_model.fittedvalues)[0,1]**2 calculates the squared correlation between the observed values of Y and the model’s predicted values. This squared correlation has a direct relationship with R-squared in a simple linear regression model, as it also represents the proportion of the variation in Y that is explained by the model. In this context, the squared correlation between Y and the predictions gives us the same R-squared value because:

A high correlation between Y and the predicted values means that the model’s predictions closely follow the pattern of the actual Y values.
Squaring the correlation coefficient removes negative values and aligns it with the interpretation of R-squared as a proportion.

4. Interpretation of np.corrcoef(Y, x)[0,1]**2

For np.corrcoef(Y, x)[0,1]**2, this expression captures the squared correlation between Y and the predictor variable, x. This squared correlation measures the proportion of variation in Y directly explained by a simple linear relationship with x, without fitting a full regression model. In other words:

This squared correlation tells us how much of the variation in Y can be explained by just looking at its linear association with x.
In simple linear regression, this value is equivalent to R-squared because the model only has one predictor. Therefore, the correlation between Y and x (when squared) provides a measure of how much Y varies with x, which is exactly what the model seeks to capture.

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import seaborn as sns

# Load the Old Faithful dataset
old_faithful = sns.load_dataset('geyser')

# Define long wait times condition
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit
long_wait_data = old_faithful[long_wait]

# Original model for long wait times
original_model = smf.ols('duration ~ waiting', data=long_wait_data).fit()
print(original_model.summary().tables[1])

# Visualize original data with trendline
fig = px.scatter(long_wait_data, x='waiting', y='duration',
                 title="Old Faithful Geyser Eruptions for long wait times (>" + str(long_wait_limit) + ")",
                 trendline='ols')
fig.show(renderer="png")

# Step 1: Bootstrap sampling distribution of slope coefficients
n_bootstraps = 1000
bootstrap_slopes = []

for _ in range(n_bootstraps):
    # Resample with replacement
    bootstrap_sample = long_wait_data.sample(n=160, replace=True)
    # Fit model and record slope
    bootstrap_model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    bootstrap_slopes.append(bootstrap_model.params['waiting'])

# Visualize the bootstrap sampling distribution of the slope coefficients
fig = px.histogram(bootstrap_slopes, nbins=30, title="Bootstrap Sampling Distribution of Slope Coefficients")
fig.update_layout(xaxis_title="Slope Coefficient", yaxis_title="Frequency")
fig.show(renderer="png")

# Step 2: Simulate sampling distribution under null hypothesis
np.random.seed(0)
n_simulations = 1000
null_slope_simulations = []

# Parameters under null hypothesis
beta0 = 1.65
beta1 = 0  # No association
sigma = 0.37

for _ in range(n_simulations):
    # Generate Y using the null hypothesis parameters
    Y_null = beta0 + beta1 * long_wait_data['waiting'] + np.random.normal(0, sigma, size=len(long_wait_data))
    simulated_data = pd.DataFrame({'waiting': long_wait_data['waiting'], 'duration': Y_null})
    # Fit model and record slope
    null_model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    null_slope_simulations.append(null_model.params['waiting'])

# Visualize the null hypothesis sampling distribution of the slope coefficients
fig = px.histogram(null_slope_simulations, nbins=30,
                   title="Sampling Distribution of Slope Coefficients Under Null Hypothesis")
fig.update_layout(xaxis_title="Slope Coefficient", yaxis_title="Frequency")
fig.show(renderer="png")

# Step 3: Check if 0 is within the 95% confidence interval of the bootstrap distribution
bootstrap_ci_lower = np.percentile(bootstrap_slopes, 2.5)
bootstrap_ci_upper = np.percentile(bootstrap_slopes, 97.5)
print(f"95% Bootstrap Confidence Interval for Slope: ({bootstrap_ci_lower}, {bootstrap_ci_upper})")

# Check if 0 is in the bootstrap CI and compare with p-value
if bootstrap_ci_lower <= 0 <= bootstrap_ci_upper:
    print("0 is within the 95% bootstrap confidence interval for the slope.")
else:
    print("0 is NOT within the 95% bootstrap confidence interval for the slope.")

# Compare with original p-value
original_p_value = original_model.pvalues['waiting']
print(f"Original model p-value for the slope: {original_p_value}")
````

````python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
from scipy import stats

# Load the Old Faithful dataset
old_faithful = sns.load_dataset('geyser')

# Define "short" and "long" wait times based on threshold
threshold = 68
old_faithful['kind'] = np.where(old_faithful['waiting'] < threshold, 'short', 'long')

# Calculate observed difference in means
mean_duration_short = old_faithful[old_faithful['kind'] == 'short']['duration'].mean()
mean_duration_long = old_faithful[old_faithful['kind'] == 'long']['duration'].mean()
observed_diff = mean_duration_long - mean_duration_short

print(f"Observed difference in means (long - short): {observed_diff}")

# Part (A): Permutation Test
n_permutations = 1000
permuted_diffs = []

for _ in range(n_permutations):
    # Shuffle the 'kind' labels
    shuffled_kind = np.random.permutation(old_faithful['kind'])
    # Calculate mean durations for shuffled groups
    mean_short = old_faithful[shuffled_kind == 'short']['duration'].mean()
    mean_long = old_faithful[shuffled_kind == 'long']['duration'].mean()
    # Store the difference
    permuted_diffs.append(mean_long - mean_short)

# Calculate p-value for the observed difference
p_value_perm = np.mean([abs(diff) >= abs(observed_diff) for diff in permuted_diffs])
print(f"Permutation test p-value: {p_value_perm}")

# Plot the permutation test results
import plotly.express as px
fig_perm = px.histogram(permuted_diffs, nbins=30, title="Permutation Test Distribution of Mean Differences")
fig_perm.add_vline(x=observed_diff, line_dash="dash", line_color="red", annotation_text="Observed Difference")
fig_perm.update_layout(xaxis_title="Mean Difference (Long - Short)", yaxis_title="Frequency")
fig_perm.show(renderer="png")

# Part (B): Bootstrap Confidence Interval
n_bootstraps = 1000
bootstrap_diffs = []

for _ in range(n_bootstraps):
    # Bootstrap samples for each group
    sample_short = old_faithful[old_faithful['kind'] == 'short']['duration'].sample(frac=1, replace=True)
    sample_long = old_faithful[old_faithful['kind'] == 'long']['duration'].sample(frac=1, replace=True)
    # Calculate and store the difference in means
    bootstrap_diffs.append(sample_long.mean() - sample_short.mean())

# Calculate the 95% confidence interval
ci_lower, ci_upper = np.quantile(bootstrap_diffs, [0.025, 0.975])
print(f"95% Bootstrap Confidence Interval for the Difference in Means: ({ci_lower}, {ci_upper})")

# Plot the bootstrap results
fig_bootstrap = px.histogram(bootstrap_diffs, nbins=30, title="Bootstrap Distribution of Mean Differences")
fig_bootstrap.add_vline(x=observed_diff, line_dash="dash", line_color="red", annotation_text="Observed Difference")
fig_bootstrap.update_layout(xaxis_title="Mean Difference (Long - Short)", yaxis_title="Frequency")
fig_bootstrap.show(renderer="png")

# Part (C): Indicator Variable Model
indicator_model = smf.ols('duration ~ C(kind, Treatment(reference="short"))', data=old_faithful).fit()
print("Indicator variable model summary:")
print(indicator_model.summary().tables[1])

# Extract p-value and confidence interval for the indicator variable
p_value_indicator = indicator_model.pvalues['C(kind, Treatment(reference="short"))[T.long]']
ci_indicator = indicator_model.conf_int().loc['C(kind, Treatment(reference="short"))[T.long]']
print(f"Indicator variable model p-value: {p_value_indicator}")
print(f"Indicator variable model 95% Confidence Interval: {ci_indicator}")
````

**(a) Explanation of Sampling Approaches**

Permutation Test: The permutation test assumes that, under the null hypothesis, the group labels "short" and "long" have no actual impact on duration. By randomly reassigning labels across the pooled data, we simulate what the distribution of differences in mean duration would look like if there were truly no difference between groups.
Bootstrap Confidence Interval: The bootstrap method assumes that each observed group is representative of its population. By resampling within each group, we approximate the sampling distribution of the mean duration difference under the observed data. This approach estimates the range of plausible differences in means based on the observed samples. Both approaches are non-parametric and data-driven, but they use different assumptions and resampling techniques to assess the difference in group means.

**(b) Comparison with the Indicator Variable Model Approach**

The indicator variable model approach (using duration ~ C(kind)) differs in several key ways but shares some conceptual similarities.

**Similarity:**

All three methods aim to assess whether there’s a significant difference in mean duration between the "short" and "long" wait groups.
Each approach ultimately examines the difference in average eruption duration between groups, either by resampling, shuffling, or using regression.
Differences:

**Methodology:**

The permutation test and bootstrap approaches involve resampling techniques that directly estimate sampling distributions without making assumptions about normality or relying on a specific statistical model.
The indicator variable model uses a linear regression framework, estimating the difference between groups based on the fitted slope (contrast) for the indicator variable. This assumes a linear relationship and uses parametric estimation methods.

**Interpretation:**

The permutation test provides a p-value that reflects the likelihood of observing a difference as extreme as the one in the data under the null hypothesis of no difference.
The bootstrap method yields a confidence interval that shows the range of plausible values for the difference in means. If this interval does not contain zero, it suggests a significant difference.
The indicator variable model estimates the difference in means and provides a p-value based on the regression coefficient for the indicator variable. This p-value tests the null hypothesis that there is no difference between groups, interpreted similarly to the permutation test p-value.

**Statistical Assumptions:**

The permutation test assumes that any observed difference is due to random label assignment under the null hypothesis.The bootstrap approach assumes that each group’s data distribution represents its population distribution.The indicator model assumes linearity, normally distributed errors, and homoscedasticity, making it more model-dependent than the non-parametric permutation and bootstrap tests.
