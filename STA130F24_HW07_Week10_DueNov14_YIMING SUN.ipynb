### Q1.1 Simple Linear Regression vs. Multiple Linear Regression

### **Simple Linear Regression**

- **Definition:** Models the relationship between a single independent (predictor) variable and a dependent (outcome) variable.
- **Equation:**
  $$ Y = \beta_0 + \beta_1X + \epsilon $$
- **Use Case:** When only one predictor is believed to influence the outcome.

### **Multiple Linear Regression**

- **Definition:** Extends simple linear regression by incorporating two or more independent variables to predict the dependent variable.
- **Equation:**
  $$ Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \dots + \beta_nX_n + \epsilon $$
- **Benefits Over Simple Linear Regression:**
  - **Enhanced Predictive Power:** By considering multiple factors, the model better captures the complexity of real-world data.
  - **Control for Confounding Variables:** Helps isolate the effect of each predictor by accounting for others.

---

### Q1.2 Continuous Variable vs. Indicator Variable in Simple Linear Regression

### **Continuous Variable**

- **Definition:** A variable that can take an infinite number of values within a range (e.g., age, income).
- **Role in Regression:** Represents a quantitative predictor where changes are measured on a continuous scale.
- **Equation Example:**
  $$ Y = \beta_0 + \beta_1X_{\text{continuous}} + \epsilon $$

### **Indicator Variable (Dummy Variable)**

- **Definition:** A binary variable that represents categorical data with two categories (e.g., gender: male=0, female=1).
- **Role in Regression:** Encodes qualitative predictors to include categorical distinctions in the model.
- **Equation Example:**
  $$ Y = \beta_0 + \beta_1X_{\text{indicator}} + \epsilon $$

### **Key Differences**

- **Nature of Data:** 
  - *Continuous Variables:* Quantify magnitude.
  - *Indicator Variables:* Categorize data.
- **Interpretation:** 
  - *Continuous Variables:* Coefficients reflect the change in \( Y \) per unit change in \( X \).
  - *Indicator Variables:* Coefficients reflect the difference in \( Y \) between categories.

---

### Q1.3 Introducing an Indicator Variable Alongside a Continuous Variable

### **Simple Linear Regression (Single Continuous Variable)**

- **Model Behavior:** Assumes a linear relationship between the continuous predictor and the outcome.
- **Data Expectation:** Changes in \( Y \) are solely driven by changes in the continuous \( X \).

### **Multiple Linear Regression (Continuous + Indicator Variable)**

- **Model Behavior:** Captures the linear relationship of each predictor while controlling for the other.
- **Data Expectation:** 
  - **Continuous Variable:** Effect remains as in simple regression.
  - **Indicator Variable:** Adds a shift or difference based on the category.
- **Overall Impact:** Accounts for both quantitative and categorical influences, allowing for more nuanced predictions.
- **Example Equation:**
  $$ Y = \beta_0 + \beta_1X_{\text{continuous}} + \beta_2X_{\text{indicator}} + \epsilon $$

---

### Q1.4 Adding an Interaction Between a Continuous and an Indicator Variable in Multiple Linear Regression

### **Effect of Interaction**

- **Definition:** An interaction term allows the effect of one predictor (e.g., continuous variable) on the outcome to vary depending on the level of another predictor (e.g., indicator variable).
  
### **Model Behavior**

- **Without Interaction:** Assumes the effect of the continuous variable is consistent across categories.
- **With Interaction:** Allows the slope of the continuous variable to differ between categories.

### **Linear Form with Interaction**

$$ Y = \beta_0 + \beta_1X_{\text{continuous}} + \beta_2X_{\text{indicator}} + \beta_3(X_{\text{continuous}} \times X_{\text{indicator}}) + \epsilon $$

### **Implications**

- **Enhanced Flexibility:** Captures more complex relationships where the impact of one predictor depends on another.
- **Interpretation:** 
  - \( \beta_1 \) represents the effect of the continuous variable when the indicator is 0.
  - \( \beta_3 \) shows how the effect of the continuous variable changes when the indicator is 1.

---

### Q1.5 Multiple Linear Regression with Indicator Variables from a Non-Binary Categorical Variable

### **Model Behavior**

- **Based Only on Indicators:** Uses multiple binary (0/1) variables to represent each category of a non-binary categorical variable.
- **Expected Data Nature:** Captures the distinct effect of each category on the outcome without assuming any inherent order or magnitude.

### **Linear Form**

$$ Y = \beta_0 + \beta_1D_1 + \beta_2D_2 + \dots + \beta_{k-1}D_{k-1} + \epsilon $$

*(Assuming \( k \) categories, with one category as the reference.)*

### **Binary Variable Encodings**

- **Dummy Coding:** Each category is represented by a separate binary variable, except one which serves as the baseline.
- **Example:** For a categorical variable with three categories (A, B, C):
  - \( D_1 = 1 \) if category A, else 0
  - \( D_2 = 1 \) if category B, else 0
  - *(Category C is the reference with both \( D_1 \) and \( D_2 \) = 0)*

### **Implications**

- **No Multicollinearity:** By omitting one category, the model avoids the dummy variable trap.
- **Interpretation:** Coefficients represent the difference in \( Y \) between each category and the reference category.

## **Q2.2. Considering Interaction Effects**

For example, a high online budget might amplify the impact of TV ads, or significant TV spending might enhance the effectiveness of online ads.

## **Q2.3. Linear Regression Models: With and Without Interaction**

### **a. Model Without Interaction**

**Equation:**
$$
Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \epsilon
$$

**Components:**
- \( Y \): Sales Revenue
- \( \beta_0 \): Intercept (baseline sales when both budgets are zero)
- \( \beta_1 \): Effect of TV Advertising Budget on Sales
- \( \beta_2 \): Effect of Online Advertising Budget on Sales
- \( \epsilon \): Error term

**Interpretation:**
- Assumes that the effect of TV advertising on sales is constant, regardless of the online advertising budget, and vice versa.

### **b. Model With Interaction**

**Equation:**
$$
Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \beta_3(X_1 \times X_2) + \epsilon
$$

**Components:**
- \( \beta_3 \): Coefficient for the interaction term between TV and Online Advertising Budgets.

**Interpretation:**
- The interaction term (\( X_1 \times X_2 \)) allows the effect of TV advertising on sales to vary depending on the level of online advertising spending, and vice versa.

---

## **Q2.4. Making Predictions: With vs. Without Interaction**

### **a. Without Interaction**

- **Prediction Formula:**
  $$
  \hat{Y} = \beta_0 + \beta_1X_1 + \beta_2X_2
  $$

- **Prediction Process:**
  1. Input the TV and Online advertising budgets (\( X_1 \) and \( X_2 \)).
  2. Multiply each by their respective coefficients (\( \beta_1 \) and \( \beta_2 \)).
  3. Add the intercept (\( \beta_0 \)) to obtain the predicted sales (\( \hat{Y} \)).

- **Nature of Predictions:**
  - Assumes additive effects of TV and Online budgets.
  - Does not account for any synergistic or antagonistic interactions between the two advertising mediums.

### **b. With Interaction**

- **Prediction Formula:**
  $$
  \hat{Y} = \beta_0 + \beta_1X_1 + \beta_2X_2 + \beta_3(X_1 \times X_2)
  $$

- **Prediction Process:**
  1. Input the TV and Online advertising budgets (\( X_1 \) and \( X_2 \)).
  2. Calculate the interaction term (\( X_1 \times X_2 \)) and multiply by \( \beta_3 \).
  3. Multiply each predictor by their respective coefficients (\( \beta_1 \) and \( \beta_2 \)).
  4. Add all terms together with the intercept (\( \beta_0 \)) to obtain the predicted sales (\( \hat{Y} \)).

- **Nature of Predictions:**
  - Captures the combined effect of TV and Online budgets.
  - Allows the impact of one advertising medium to change based on the level of the other, providing a more nuanced and potentially accurate prediction.

### **High-Level Differences Between Models**

- **Additive Model (Without Interaction):**
  - **Simplicity:** Easier to interpret as each predictor's effect is independent.
  - **Limitation:** May miss important synergistic effects, leading to biased or inaccurate predictions if interactions are indeed present.

- **Interactive Model (With Interaction):**
  - **Complexity:** More complex due to the additional interaction term.
  - **Advantage:** Provides a more flexible and potentially accurate representation by accounting for the interdependence between predictors.

---

### Q4.1. R-squared: Explaining the Overall Model Fit

**R-squared** is a measure of how well the model explains the variance in the dependent variable (in this case, **HP**). It represents the proportion of the variance in **HP** that is predictable from the independent variables (**Sp. Def** and **Generation** with interaction terms).

- **R-squared = 0.176 (17.6%)** means that only 17.6% of the variability in **HP** is explained by the model. 
- **Low R-squared** indicates that a large portion (82.4%) of the variability in **HP** is due to factors not included in the model. This might be because other important predictors, such as **Attack**, **Defense**, or **Speed**, are not part of the model.

**Implications of Low R-squared:**
- The model does not capture all factors influencing **HP**.
- There may be other variables that have substantial effects on **HP** which are not included in this model.
- **R-squared** focuses on the overall goodness of fit of the model and does not tell us about the importance or significance of individual predictors.

---

### Q4.2. Significant Coefficients: Strong Evidence for Individual Predictors

While the overall R-squared is low, **many of the coefficients are statistically significant** and have relatively large values. Each coefficient represents the change in **HP** for a one-unit change in the predictor variable, holding all other variables constant.

- **Statistical Significance (Low p-values):** The p-values for many coefficients are below 0.05, indicating strong evidence against the null hypothesis of "no effect" for these predictors.

- **Significant predictors** indicate reliable associations with **HP** but do not imply that they explain a large proportion of the total variability.

The low R-squared and significant coefficients can coexist without contradiction because:

1. **R-squared assesses the overall fit**, which evaluates how much of the variance in **HP** is captured by the model.

## Q6.1. Design Matrix in Model 4

The **model4_linear_form** specifies a complex regression model that includes multiple predictors (**Attack**, **Defense**, **Speed**, **Legendary**, **Sp. Def**, and **Sp. Atk**) along with interaction terms among them. This specification creates a large number of new predictor variables in the design matrix, which is represented by **model4_spec.exog**.

### Design Matrix Explained

- **Design Matrix (model4_spec.exog)**: This matrix contains the transformed and expanded set of predictor variables used to fit the regression model. Each column represents a predictor or interaction term, and each row corresponds to an observation in the dataset.
  
- **Shape of the Design Matrix**: The shape of **model4_spec.exog** indicates the number of rows (observations) and columns (predictor variables). The high number of columns in this matrix reflects the complexity introduced by multiple interaction terms.

The design matrix captures the expanded set of variables created from the linear form specification, allowing the model to account for interactions between predictors. However, this complexity comes with potential issues, particularly multicollinearity.

---

## Q6.2. Multicollinearity in the Design Matrix

### What is Multicollinearity?

- **Multicollinearity** occurs when two or more predictor variables in the design matrix are highly correlated. This means that some predictors can be nearly linear combinations of others, leading to redundancy and instability in the model.

### Causes of Multicollinearity in Model 4

- **Interaction Terms**: Including many interaction terms increases the likelihood of multicollinearity, as interactions between similar or related predictors can result in highly correlated columns in the design matrix.
  
- **High Condition Number**: The "Condition Number" in the model summary is a measure of multicollinearity. A high condition number (e.g., 12,000,000,000,000,000) suggests severe multicollinearity in **model4_fit**. This high condition number indicates that some columns in the design matrix are nearly linearly dependent.

### Impact of Multicollinearity on Out-of-Sample Performance

- **Unstable Coefficient Estimates**: Multicollinearity makes it difficult for the model to estimate reliable coefficients, as small changes in the data can lead to large variations in these estimates.
  
- **Poor Generalization**: Models with high multicollinearity often perform well on the training data but poorly on unseen data. This is because the model has "overfit" to noise or redundant patterns in the training data, reducing its ability to generalize to new data.

Multicollinearity in the design matrix thus contributes to the lack of out-of-sample generalization observed in **model4_fit**.

---

## Q6.3. Centering and Scaling to Reduce Multicollinearity

### Purpose of Centering and Scaling

Centering and scaling (standardizing) the predictors can help reduce multicollinearity by transforming the predictor variables to have a mean of zero and a standard deviation of one. This transformation makes the design matrix more stable and improves the condition number.

- **Centering**: Shifts the predictor values so that their mean is zero.
- **Scaling**: Adjusts the predictor values so that their standard deviation is one.

### Effect on the Condition Number

- **Initial Condition Number**: In **model3_fit** (a simpler model with only **Attack** and **Defense**), the condition number was high (343.0) without centering and scaling, indicating moderate multicollinearity.
  
- **Reduced Condition Number**: After centering and scaling, the condition number for the centered and scaled version of **model3_fit** dropped to 1.66, a very low and stable value.

### Impact on Model Stability

By centering and scaling the predictors, we reduce multicollinearity, leading to:

- **More Stable Coefficient Estimates**: The coefficients become less sensitive to small changes in the data, improving model robustness.
- **Improved Generalizability**: Reducing multicollinearity helps the model generalize better to new data, as it minimizes overfitting to redundant patterns in the training data.
